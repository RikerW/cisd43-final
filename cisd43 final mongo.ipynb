```python
import pymongo, pandas as pd
```

```python
# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
client
```

```python
db = client.housing
db
```

```python
collection = db.house
collection
```

```python
# Retrieve the first three records
records = collection.find({}).limit(3)

# Print the records
for rec in records:
    print("{")
    for key in rec.keys():
        print(f"{key}: {rec[key]},")
    print("},")
```

#### Find one document with invoice number is equals to 536365

```python
# Find one document
result = collection.find({"price" : 221900}).limit(1)

for rec in result:
    print("{")
    for key in rec.keys():
        print(f"{key}: {rec[key]},")
    print("},")
```
